```python
import matplotlib.pyplot as plt

sizes = []
times = []

with open("files/result.txt", "r") as file:
    for line in file:
        parts = line.split()
        sizes.append(int(parts[1]))
        times.append(float(parts[3]))

plt.figure(figsize=(10, 4), dpi=250)
plt.plot(sizes, times, marker='o', color='green')
plt.xlabel('Size')
plt.ylabel('Time, s')
plt.title('Time vs Size')
plt.grid()
plt.savefig('files/plot.png')
plt.show()
```
